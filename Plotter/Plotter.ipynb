```python
def createdata(data):
    if(data == 1):
        x = np.random.rand(100,5)
        df1 = pd.DataFrame(x, columns=['A', 'B', 'C', 'D', 'E'])
    elif(data == 2):
        x  = [0,0,0,0,0]
        r1 = [0,0,0,0,0]
        r2 = [0,0,0,0,0]
        r3 = [0,0,0,0,0]
        r4 = [0,0,0,0,0]
        print('Insert columns values')
        print('| x | x | x | x |')
        print(' ---------------')
        print('|   |   |   |   |\n')
        i = 0
        for i in [0,1,2,3,4]:
            x[i] = input()
            i += 1
            if (i == 5):
                print('\n')
        print('Insert first row values')
        print('|   |   |   |   |')
        print(' ---------------')
        print('| 1 | 1 | 1 | 1 |\n')
        i = 0
        for i in [0,1,2,3,4]:
            r1[i] = int(input())
            i += 1
            if (i == 5):
                print('\n')
        print('Insert second row values')
        print('|   |   |   |   |')
        print(' ---------------')
        print('| 2 | 2 | 2 | 2 |\n')
        i = 0
        for i in [0,1,2,3,4]:
            r2[i] = int(input())
            i += 1
            if (i == 5):
                print('\n')
        print('Insert third row values')
        print('|   |   |   |   |')
        print(' ---------------')
        print('| 3 | 3 | 3 | 3 |\n')
        i = 0
        for i in [0,1,2,3,4]:
            r3[i] = int(input())
            i += 1
            if (i == 5):
                print('\n')
        print('Insert fourth row values')
        print('|   |   |   |   |')
        print(' ---------------')
        print('| 4 | 4 | 4 | 4 |\n')
        i = 0
        for i in [0,1,2,3,4]:
            r4[i] = int(input())
            i += 1
            if (i == 5):
                print('\n')
        df1 = pd.DataFrame([r1, r2, r3, r4], columns = x)
    elif(data == 3):
        print('Insert file name')
        file = input()
        x = pd.read_csv(file)
        df1 = pd.DataFrame(x)
        print(file, ' Loaded!!')
    else:
        print('DataFrame creation failed, please insert a different number')
        return
    return df1
```

```python
def plotter(plot):
    if (plot == 1):
        finalplot = df1.iplot(kind='scatter')
    elif (plot == 2):
        finalplot = df1.iplot(kind='scatter', mode='markers', symbol='x', colorscale='paired')
    elif (plot == 3):
        finalplot = df1.iplot(kind='bar')
    elif (plot == 4):
        finalplot = df1.iplot(kind='hist')
    elif (plot == 5):
        finalplot = df1.iplot(kind='box')
    elif (plot == 6):
        finalplot = df1.iplot(kind='surface')
    else:
        finalplot = print('Please insert a valid option')
    return finalplot    
```

```python
def plotter2(plot):
    col = input('how many columns do you want to plot (1-3): ')
    col = int(col)
    if (col == 1):
        print('\n', list(df1))
        colm = input('Insert column name you want to plot?: ')
        if (plot == 1):
            finalplot = df1[colm].iplot(kind='scatter')
        elif (plot == 2):
            finalplot = df1[colm].iplot(kind='scatter', mode='markers', symbol='x', colorscale='paired')
        elif (plot == 3):
            finalplot = df1[colm].iplot(kind='bar')
        elif (plot == 4):
            finalplot = df1[colm].iplot(kind='hist')
        elif (plot == 5):
            finalplot = df1[colm].iplot(kind='box')
        elif (plot == 6 or plot == 7):
            finalplot = print('Blubble or Surface plto require more than one column :c')
        else:
            finalplot = print('Invalid plotting selection')
    elif (col == 2):
        print('\n', list(df1))
        print('Insert columns names you want to plot')
        x = input('First column: ')
        y = input('Second column: ')
        if (plot == 1):
            finalplot = df1[[x,y]].iplot(kind='scatter')
        elif (plot == 2):
            finalplot = df1[[x,y]].iplot(kind='scatter', mode='markers', symbol='x', colorscale='paired')
        elif (plot == 3):
            finalplot = df1[[x,y]].iplot(kind='bar')
        elif (plot == 4):
            finalplot = df1[[x,y]].iplot(kind='hist')
        elif (plot == 5):
            finalplot = df1[[x,y]].iplot(kind='box')
        elif (plot == 6):
            finalplot = df1[[x,y]].iplot(kind='surface')
        elif (plot == 7):
            list(df1)
            size = input('Insert bubble size (other column): ')
            finalplot = df1.iplot(kind='bubble', x = x, y = y, size = size)
        else:
            finalplot = print('Invalid plotting selection')
    elif (col == 3):
        print('\n', list(df1))
        print('Insert columns names you want to plot')
        x = input('First column: ')
        y = input('Second column: ')
        z = input('Third column: ')
        if (plot == 1):
            finalplot = df1[[x,y,z]].iplot(kind='scatter')
        elif (plot == 2):
            finalplot = df1[[x,y,z]].iplot(kind='scatter', mode='markers', symbol='x', colorscale='paired')
        elif (plot == 3):
            finalplot = df1[[x,y,z]].iplot(kind='bar')
        elif (plot == 4):
            finalplot = df1[[x,y,z]].iplot(kind='hist')
        elif (plot == 5):
            finalplot = df1[[x,y,z]].iplot(kind='box')
        elif (plot == 6):
            finalplot = df1[[x,y,z]].iplot(kind='surface')
        elif (plot == 7):
            list(df1)
            print('\n', list(df1))
            size = input('Insert bubble size (other column): ')
            finalplot = df1.iplot(kind='bubble', x = x, y = y, z = z, size = size)
        else:
            finalplot = print('Invalid plotting selection')
    else:
        finalplot = print('Please insert a valid option c:')
    
    return finalplot 
```

```python
def main(cat):
    if (cat == 1):
        print('Select plot style (1-6)\n')
        print('\t\033[35m(1)\033[0m Line plot')
        print('\t\033[35m(2)\033[0m Scatter plot')
        print('\t\033[35m(3)\033[0m Bar plot')
        print('\t\033[35m(4)\033[0m histogram')
        print('\t\033[35m(5)\033[0m Box plot')
        print('\t\033[35m(6)\033[0m Surface plot')
        plot = int(input())
        output = plotter(plot)
    elif (cat == 2):
        print('Select plot style (1-7)\n')
        print('\t\033[35m(1)\033[0m Line plot')
        print('\t\033[35m(2)\033[0m Scatter plot')
        print('\t\033[35m(3)\033[0m Bar plot')
        print('\t\033[35m(4)\033[0m histogram')
        print('\t\033[35m(5)\033[0m Box plot')
        print('\t\033[35m(6)\033[0m Surface plot')
        print('\t\033[35m(7)\033[0m Bubble plot')
        plot = int(input())
        output = plotter2(plot)
    else:
        print('Please insert a valid option (1-2)')
```
